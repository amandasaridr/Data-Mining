# Metode K-Means (Elbow)

```python
import numpy as np
import matplotlib.pyplot as plt

#mengimport dataset
data = df_hepatitis
#mengubah datasetnya menjadi bentuk numpy array
a = data.values
class KMeans:
    def __init__(self, n_clusters, max_iter=300):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
    def fit(self, a):
        n_samples, n_features = a.shape
        #menginisialisasi centroid secara acak
        self.centroids = a[np.random.choice(n_samples, self.n_clusters, replace=False)]
        #mengiterasi untuk mengoptimalkan centroid
        for _ in range(self.max_iter):
            #menghitung jarak setiap titik data ke setiap centroid
            jarak_titik = np.sqrt(((a - self.centroids[:, np.newaxis])**2).sum(axis=2))
            #memperbarui label cluster untuk setiap titik data
            label_cluster = np.argmin(jarak_titik, axis=0)
            #memperbarui centroid dengan rata-rata titik data dalam setiap cluster
            centroid_baru = np.array([a[label_cluster == f].mean(axis=0) for f in range(self.n_clusters)])
            #jika tidak ada perubahan dalam centroid, keluar dari iterasi
            if np.allclose(self.centroids, centroid_baru):
                break
            self.centroids = centroid_baru
        #menghitung inertia 
        self.inertia = 0
        for f in range(self.n_clusters):
            point_cluster = a[label_cluster == f]
            self.inertia += np.sum((point_cluster - self.centroids[f])**2)
        return self.inertia
#menentukan jumlah cluster optimal menggunakan metode Elbow
def find_optimal_clusters(a, max_clusters):
    inertias = []
    for n_clusters in range(1, max_clusters+1):
        kmeans = KMeans(n_clusters=n_clusters)
        inertia = kmeans.fit(a)
        inertias.append(inertia)
    return inertias
#contoh penerapan k-means:
if __name__ == "__main__":
    max_clusters = 10
    inertias = find_optimal_clusters(a, max_clusters)
    print("Hasil nilai dari k-means dari data hepatitis\n", inertias)
    #membuat plot metode elbow
    plt.plot(range(1, max_clusters+1), inertias, marker='o')
    plt.title("metode elbow hepatitis")
    plt.xlabel("jumlah cluster")
    plt.ylabel("nilai k-means atau inertia")
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import pandas as pd

#mengimport dataset
data = df_hepatitis
#mengubah datasetnya menjadi bentuk numpy array
a = data.values
class KMeans:
    def __init__(self, n_clusters, max_iter=300, random_state=None):
        self.n_clusters = n_clusters
        self.max_iter = max_iter
        self.random_state = random_state
    def fit(self, a):
        n_samples, n_features = a.shape
        #menginisialisasi seed acak (jika random_state diberikan)
        if self.random_state is not None:
            np.random.seed(self.random_state)
        #menginisialisasi centroid secara acak
        self.centroids = a[np.random.choice(n_samples, self.n_clusters, replace=False)]
        #mengiterasi untuk mengoptimalkan centroid
        for _ in range(self.max_iter):
            #menghitung jarak setiap titik data ke setiap centroid
            jarak_titik = np.sqrt(((a - self.centroids[:, np.newaxis])**2).sum(axis=2))
            #memperbarui label cluster untuk setiap titik data
            label_cluster= np.argmin(jarak_titik, axis=0)
            #memperbarui centroid dengan rata-rata titik data dalam setiap cluster
            centroid_baru = np.array([a[label_cluster == f].mean(axis=0) for f in range(self.n_clusters)])
            #jika tidak ada perubahan dalam centroid, keluar dari iterasi
            if np.allclose(self.centroids, centroid_baru):
                break
            self.centroids = centroid_baru
    def predict(self, a):
        #menghitung jarak setiap titik data ke setiap centroid
        jarak_titik = np.sqrt(((a - self.centroids[:, np.newaxis])**2).sum(axis=2))
        #mengembalikan label cluster dengan jarak terdekat
        return np.argmin(jarak_titik, axis=0)
#contoh penerapan:
if __name__ == "__main__":
    #menginisialisasi dan latih model K-Means dengan n cluster
    kmeans = KMeans(n_clusters=4, random_state=42)
    kmeans.fit(a)
    #memprediksi label cluster untuk setiap titik data
    label_prediksi = kmeans.predict(a)
    print("Hasil dari prediksi label cluster\n",label_prediksi)
#tambahkan outputnya kedalam df dan tampilkan
df_hepatitis["Cluster"] = label_prediksi
dataset = pd.DataFrame(df_hepatitis)
print("\nTabel data hepatitis setelah dimasukkan kolom cluster")
dataset
```

# Silhouette Coefficient

```python
import numpy as np

def silhouette_coefficient(data, label_prediksi):
    n_samples = len(data)
    silhouette_scores = []
    #mengiterasi melalui setiap sampel
    for c in range(n_samples):
        #mengambil label cluster untuk sampel c
        cluster_label = label_prediksi[c]
        #menghitung a(c): rata-rata jarak dari sampel c ke sampel yang lain dalam cluster yang sama
        samples_in_cluster = dataset[label_prediksi == cluster_label]
        ac = np.mean(np.linalg.norm(data[c] - samples_in_cluster, axis=1))
        #menghitung b(c): jarak rata-rata dari sampel c ke semua sampel yang lain dalam cluster lain terdekat
        other_clusters = np.unique(label_prediksi[label_prediksi != cluster_label])
        value_bc = []
        for other_cluster in other_clusters:
            samples_in_other_cluster = dataset[label_prediksi == other_cluster]
            value_bc.append(np.mean(np.linalg.norm(dataset[c] - samples_in_other_cluster, axis=1)))
        bc = min(value_bc)
        #menghitung silhouette coefficient untuk sampel c
        silhouette_coefficient = (bc - ac) / max(ac, bc)
        silhouette_scores.append(silhouette_coefficient)
    #menghitung silhouette coefficient rata-rata dari semua sampel
    rata2_silhouette_coefficient = np.mean(silhouette_scores)
    return rata2_silhouette_coefficient
#data hepatitis
dataset = df_hepatitis.values
label_prediksi = df_hepatitis.Cluster.values
#menghitung silhouette coefficient
rata2_silhouette_coefficient = silhouette_coefficient(dataset, label_prediksi)
print("Hasil dari silhouette coefficient:", rata2_silhouette_coefficient)
#membuat plot silhouette coefficient untuk setiap sampel
silhouette_values = silhouette_coefficient(dataset, label_prediksi)
plt.figure(figsize=(8, 6))
plt.hist(silhouette_values, bins=20, color="lightblue", edgecolor="red")
plt.xlabel("nilai silhouette coefficient")
plt.ylabel("frekuensi")
plt.title("silhouette coefficient")
plt.grid(True)
plt.show()
```
